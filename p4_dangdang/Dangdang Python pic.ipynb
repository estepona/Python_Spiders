先建一个url_list

```python
url_list = []
for i in range(1,20):
    url_list.append(url + str(i))
url_list
```

item里的图片规律是<img data-original 并含有style

```python
# 先测试第一页的
p1 = BS(requests.get('http://search.dangdang.com/?key=Python&act=input&page_index=1').content, 'html.parser')
```

```python
p1_img_url = []
for i in p1.find_all('img'):
    if i.has_attr('data-original'):
        p1_img_url.append(i.get('data-original'))
p1_img_url
```

```python
# create a folder named "pic"
os.mkdir('.\pic')
```

```python
for i in range(len(p1_img_url)):
    img = requests.get(p1_img_url[i]).content
    with open('.\pic\%d.jpg' % (i+1), 'wb') as handler:
        handler.write(img)
```

```python
len(p1_img_url)
```

做整体的loop和download

```python
img_url = []
for i in url_list:
    p = BS(requests.get(i).content, 'html.parser')
    for j in p.find_all('img'):
        if j.has_attr('data-original'):
            img_url.append(j.get('data-original'))
img_url
```

```python
for i in range(len(img_url)):
    img = requests.get(img_url[i]).content
    with open('.\pic\%d.jpg' % (i+1), 'wb') as handler:
        handler.write(img)
```

把所有图书的名字放入一个列表: 找a里的title

```python
title_list = []
for i in url_list:
    p = BS(requests.get(i).content.decode('gb2312', 'ignore'), 'html.parser') # BS直接解码
    for j in p.find_all('a', {'class':'pic'}):
        title_list.append(j.get('title'))
title_list
```

```python
len(title_list)
# 长度不一致，应该在一开始就把title和图片放在一起
```
